```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(columns=[i for i in df.columns if df[i].isna().sum()>800],axis=1)
df
```

```python
df.info()
```

```python
for col in df.columns:
  if df[col].dtype in [np.int64, np.float64]:
    df[col].fillna(df[col].mean(), inplace=True)
  else:
    df[col].fillna(method='ffill', inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
  if df[col].dtype == object:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R2 score:", r2*100)
```

```python
import matplotlib.pyplot as plt
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='solid', lw=2, label='Predicted')
plt.scatter(y_test, y_pred, color='red', marker='o', label='Actual')
plt.title('Linear Regression Model(Actual VS Predicted)')
plt.legend()
plt.show()
```
